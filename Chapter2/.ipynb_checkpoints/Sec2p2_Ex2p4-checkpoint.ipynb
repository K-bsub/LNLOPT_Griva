## Problem Statement

Consider the problem

minimize $f(x) = x_{1} \\$
subject to $(x_{1}-1)^{2} + x_{2}^{2} = 1 $
and $(x_{1}+1)^{2} + x_{2}^{2} = 1 $

Graph the feasible set and locate the stationary points, local minima, and global minima.

## Import packages

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import copy as cp
```

## Evaluate function

Let's evaluate the value of the function for values for x from -1000 to 1000. 

```python
def f(x, y):
    return x

x1 = np.arange(-5000, 5001, 1)
x2 = np.arange(-5000, 5001, 1)
X, Y = np.meshgrid(x1, x2)
Z = f(X, Y)
```

## Find the feasible set

```python
const1 = ((X-1)**2 + Y**2 == 1)
const2 = ((X+1)**2 + Y**2 == 1)
x1_f = X[const1 & const2]
x2_f = Y[const1 & const2]
f_f = Z[const1 & const2]
f_f.shape
```

## Plot the function

Let's use Matplotlib to plot the function versus x, and indicate the stationary points, local minima, and the global minima.

```python
fig = plt.figure(figsize=(12, 7))
ax = plt.axes(projection='3d')
#ax.contour3D(X, Y, Z, 100)
ax.scatter3D(x1_f, x2_f, f_f, color='red');
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f')
ax.set_title('Feasible Set');
ax.view_init(5, 50)
```

Based on the plot, there is a **global minimizer** at $f = -0.01$ and $x_{1} and x_{2} = -0.05$.
