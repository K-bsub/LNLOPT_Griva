## Problem Statement

Let $x_k = (2,1)^{T}$ and $p_k = (-1,3)^{T}$. Plot the set $\{x:x = x_k + \alpha p_k, \alpha \ge 0\}$.

## Import packages

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Define $x_k$ and $p_k$

```python
xk = np.array([ 2, 1])
pk = np.array([-1, 3])
```

## Define $\alpha$

Let's define the range of $\alpha$ to be from 0 to 10.

```python
alpha = np.arange(0, 10.1, 0.1)
alpha.shape
```

## Calculate x

```python
x = xk + alpha[:,None]*pk
x.shape
```

## Plot x

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
ax.plot(x[:,0], x[:,1], color="#8C7298")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
