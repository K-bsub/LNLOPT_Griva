## Problem Statement

Consider the one-variable function

$$
f(x) = (x + 1)x(x - 2)(x - 5)
$$

Graph this function and locate the stationary points, local minima, and global minima.

## Import packages

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Evaluate function

Let's evaluate the value of the function for values for x from -1000 to 1000. 

```python
X = [*range(-1000,1001,1)]
f = []
for x in X:
    f.append((x+1)*x*(x-2)*(x-5))
```

## Plot the function

Let's use Matplotlib to plot the function versus x, and indicate the stationary points, local minima, and the global minima.

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
ax.plot(X, f, color="#8C7298")
ax.text(-240, 1e12, r"$f(x)=(x+1)x(x-2)(x-5)$")
ax.annotate("",(-250,0.025e12),(250,0.025e12),arrowprops=dict(arrowstyle='<->'))
ax.text(-340,0.04e12,'nonstrict global minimizer & stationary points')
plt.xlabel('x')
plt.ylabel('f')
plt.show()
```
